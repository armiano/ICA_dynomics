```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy as sc
from scipy import stats


dfs = []
df1 = pd.read_csv("dynomics_5.0uM_Cr(VI)_background_normalized_smoothed.csv")
df2 = pd.read_csv("dynomics_5.0uM_Fe(III)_background_normalized_smoothed.csv")
df3 = pd.read_csv("dynomics_0.04uM_Cd(II)_background_normalized_smoothed.csv")
df4 = pd.read_csv("dynomics_0.765uM_Zn(II)_background_normalized_smoothed.csv")
df5 = pd.read_csv("dynomics_2.0uM_Cu(II)_background_normalized_smoothed.csv")
df6 = pd.read_csv("dynomics_0.14uM_Pb(II)_background_normalized_smoothed.csv")
dfs.extend([df1,df2,df3,df4,df5,df6])
```

```python
#in this version we have All the genes instead of a subsection and we divided the induction window in 6 subparts
#each subpart has a correspondent fold change which is calculated by dividing the given point over the point at time 36 
#corresponding to the beginning of the induction window. 


for k, df in enumerate(dfs):
    df = dfs[k]
    col_num = len(df.columns)
    df_gene_time= df[df.columns[5:len(df.columns)]]
    #df_gene_time.head(2)
    df_gene_time = df_gene_time.rename(index = df_gene_time["gene_name"])
    genes_unique = df_gene_time.index.unique()
    del df_gene_time["gene_name"]
    genes_unique = genes_unique[:-1]
    #print(genes_unique)
    
    
    #calculate fold changes using the normalized values of the GFP
    #calculate the fold change for all replicates for each gene
    #Then average the fold change
    #then take the derivative 

    df_foldChange = pd.DataFrame(columns = ['Gene_name', 'point1', 'point2', 'point3', 'point4', 'point5', 'point6']) 
    #print(df_foldChange.head(3))

    df_foldChange["Gene_name"] = genes_unique  #list of unique genes
    
    #from 0 to num unique genes in dataset
    for i in range(len(genes_unique)-1): 
            #print('caio')
            df_gene = df_gene_time.groupby(df_gene_time.index).get_group(genes_unique[i]) #get all replicates for each gene
            #foldChange =[]
           # print(df_gene)
            foldChange = np.zeros((len(df_gene),6))
            for a in range(len(df_gene)):  #loop through replicates
                point = 0
                for j in range(0,6):
                    point = 36+(j+1)*4
                    if df_gene.iloc[a, point] > 0 and df_gene.iloc[a,36] > 0:
                        foldChange[a,j] = (df_gene.iloc[a, point] / df_gene.iloc[a, 36])
                        #print(foldChange[a,j])
                    elif df_gene.iloc[a, point] < 0 and df_gene.iloc[a,36] < 0:
                        foldChange[a,j] = (df_gene.iloc[a, 36] / df_gene.iloc[a, point])
                        #print(foldChange[a,j])
                    elif df_gene.iloc[a, point] < 0 and df_gene.iloc[a,36] > 0:
                        foldChange[a,j] = (abs(df_gene.iloc[a,36] / (df_gene.iloc[a, point] - df_gene.iloc[a, 36])))
                    elif df_gene.iloc[a, point] > 0 and df_gene.iloc[a,36] < 0:
                        foldChange[a,j] = (abs((df_gene.iloc[a, point] - df_gene.iloc[a, 36])/df_gene.iloc[a,36]))
            #print(foldChange)
            #print('ciao')

            logfoldChange = np.log2(foldChange.mean(axis=0)) 
            #print(logfoldChange)

            df_foldChange.iloc[i,1] = logfoldChange[0]
            df_foldChange.iloc[i,2] = logfoldChange[1]
            df_foldChange.iloc[i,3] = logfoldChange[2]
            df_foldChange.iloc[i,4] = logfoldChange[3]
            df_foldChange.iloc[i,5] = logfoldChange[4]
            df_foldChange.iloc[i,6] = logfoldChange[5]
    #print(df_foldChange)
    #assign a value of 0 to all the NaN
    #df_foldChange['point1'] = df_foldChange['point1'].fillna(0)
    #df_foldChange['point2'] = df_foldChange['point2'].fillna(0)
    #df_foldChange['point3'] = df_foldChange['point3'].fillna(0)
    #df_foldChange['point4'] = df_foldChange['point4'].fillna(0)
    #df_foldChange['point5'] = df_foldChange['point5'].fillna(0)
    #df_foldChange['point6'] = df_foldChange['point6'].fillna(0)
    


    df_foldChange_geneidx = df_foldChange.rename(index = df_foldChange["Gene_name"])
    del df_foldChange_geneidx["Gene_name"]
    #Add a column with the pvalues previously calculated to the dataset
   # del df_foldChange_geneidx["Fold_changeFe100"]
    #df_foldChange_post_avg_pval = df_foldChange_geneidx.assign(PValues=pvalue)
    #df_foldChange_post_avg_pval
    genes_idx = []
    genes_idx = np.asarray(genes_unique)



    df_foldChange = df_foldChange.rename(index = df_foldChange["Gene_name"])

    #df_foldChange_geneidx["Pvalue"] = pvalue
    # df_foldChange_logOnly = df_foldChange_geneidx
    #del df_foldChange_logOnly["Pvalue"]
    
    #k indicates which dataset we are looking at in terms of induction
    if k == 0:
        data_all = df_foldChange
        data_all = data_all.rename(columns = {'point1' : 'Cr_1'})
        data_all = data_all.rename(columns = {'point2' : 'Cr_2'})
        data_all = data_all.rename(columns = {'point3' : 'Cr_3'})
        data_all = data_all.rename(columns = {'point4' : 'Cr_4'})
        data_all = data_all.rename(columns = {'point5' : 'Cr_5'})
        data_all = data_all.rename(columns = {'point6' : 'Cr_6'})
    elif k == 1:
        #print(df_foldChange[["point1"]].copy())
        data_all["Fe_1"] = df_foldChange[["point1"]].copy()
        data_all["Fe_2"] = df_foldChange[["point2"]].copy()
        data_all["Fe_3"] = df_foldChange[["point3"]].copy()
        data_all["Fe_4"] = df_foldChange[["point4"]].copy()
        data_all["Fe_5"] = df_foldChange[["point5"]].copy()
        data_all["Fe_6"] = df_foldChange[["point6"]].copy()
        #print(data_all["Fe_1"])
    elif k == 2:
        data_all["Cd_1"] = df_foldChange.loc[:, ["point1"]]
        data_all["Cd_2"] = df_foldChange.loc[:, ["point2"]]
        data_all["Cd_3"] = df_foldChange.loc[:, ["point3"]]
        data_all["Cd_4"] = df_foldChange.loc[:, ["point4"]]
        data_all["Cd_5"] = df_foldChange.loc[:, ["point5"]]
        data_all["Cd_6"] = df_foldChange.loc[:, ["point6"]]
    elif k == 3:       
        data_all["Zn_1"] = df_foldChange.loc[:, ["point1"]]
        data_all["Zn_2"] = df_foldChange.loc[:, ["point2"]]
        data_all["Zn_3"] = df_foldChange.loc[:, ["point3"]]
        data_all["Zn_4"] = df_foldChange.loc[:, ["point4"]]
        data_all["Zn_5"] = df_foldChange.loc[:, ["point5"]]
        data_all["Zn_6"] = df_foldChange.loc[:, ["point6"]]
    elif k == 4:    
        data_all["Cu_1"] = df_foldChange.loc[:, ["point1"]]
        data_all["Cu_2"] = df_foldChange.loc[:, ["point2"]]
        data_all["Cu_3"] = df_foldChange.loc[:, ["point3"]]
        data_all["Cu_4"] = df_foldChange.loc[:, ["point4"]]
        data_all["Cu_5"] = df_foldChange.loc[:, ["point5"]]
        data_all["Cu_6"] = df_foldChange.loc[:, ["point6"]]
    elif k ==5:
        data_all["Pb_1"] = df_foldChange.loc[:, ["point1"]]
        data_all["Pb_2"] = df_foldChange.loc[:, ["point2"]]
        data_all["Pb_3"] = df_foldChange.loc[:, ["point3"]]
        data_all["Pb_4"] = df_foldChange.loc[:, ["point4"]]
        data_all["Pb_5"] = df_foldChange.loc[:, ["point5"]]
        data_all["Pb_6"] = df_foldChange.loc[:, ["point6"]]
    
    data_all = data_all.rename(index = data_all["Gene_name"])
```
