```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Supplementary_Data_3.csv',index_col=0)

# Define a threshold (you would define a suitable value based on your data)
threshold = 0.2

# Initialize a dictionary to store the results
results = {}

# Iterate over each iModulon
for i, row in data.iterrows():
    max_amplitude = -1
    enriched_metal = None
    
    # Iterate over each heavy metal (assuming each metal corresponds to 6 columns)
    for j in range(0, data.shape[1], 6):
        # Calculate the amplitude for the current metal
        amplitude = row.iloc[j:j+6].max() - row.iloc[j:j+6].min()
        
        # Check if the amplitude exceeds the threshold and is the highest so far
        if amplitude > threshold and amplitude > max_amplitude:
            max_amplitude = amplitude
            enriched_metal = data.columns[j].split('_')[0]  # Assuming column names include metal names
    
    # Store the result for the current iModulon
    results[f'iModulon_{i+1}'] = enriched_metal if enriched_metal else 'No enrichment'

# Print the results
for iModulon, metal in results.items():
    print(f'{iModulon}: {metal}')
```

```python
#Add secondary enriched heavy metals

import pandas as pd

# Load the dataset
data = pd.read_csv('Supplementary_Data_3.csv',index_col=0)

# Define a threshold for primary and secondary metal identification
primary_threshold = 0.5
secondary_threshold = 0.5

# Initialize a dictionary to store the results
results = {}

# Iterate over each iModulon
for i, row in data.iterrows():
    max_amplitude = -1
    enriched_metal = None

    # First loop to identify the primary metal
    for j in range(0, data.shape[1], 6):
        # Calculate the amplitude for the current metal
        amplitude = row.iloc[j:j+6].max() - row.iloc[j:j+6].min()
        
        # Check if the amplitude exceeds the primary threshold and is the highest so far
        if amplitude > primary_threshold and amplitude > max_amplitude:
            max_amplitude = amplitude
            enriched_metal = data.columns[j].split('_')[0]  # Assuming column names include metal names

    # Second loop to identify the secondary metals
    secondary_metals = []
    for j in range(0, data.shape[1], 6):
        # Calculate the amplitude for the current metal
        amplitude = row.iloc[j:j+6].max() - row.iloc[j:j+6].min()
        
        # Check if the amplitude exceeds the secondary threshold and the metal is not the primary metal
        if amplitude > secondary_threshold and data.columns[j].split('_')[0] != enriched_metal:
            secondary_metals.append(data.columns[j].split('_')[0])
    
    # Store the result for the current iModulon
    results[f'iModulon_{i+1}'] = {'Primary': enriched_metal if enriched_metal else 'No enrichment', 'Secondary': ", ".join(secondary_metals) if secondary_metals else 'No secondary enrichment'}

# Print the results
for iModulon, metal in results.items():
    print(f'{iModulon}: Primary - {metal["Primary"]}, Secondary - {metal["Secondary"]}')
```

```python
# Define a dictionary to map heavy metals to their respective columns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Load the dataset
data = pd.read_csv('Supplementary_Data_3.csv')

# Define a threshold (you would define a suitable value based on your data)
threshold = 0.5

# Initialize a dictionary to store the results
results = {}

# Get the list of heavy metals from the column names
metals = [col.split('_')[0] for col in data.columns[1::6]]

# Iterate over each iModulon
for i, row in data.iterrows():
    max_amplitude = -1
    enriched_metal = None
    
    # Iterate over each heavy metal
    for j, metal in enumerate(metals):
        # Calculate the amplitude for the current metal
        amplitude = row.iloc[j*6+1:(j+1)*6+1].max() - row.iloc[j*6+1:(j+1)*6+1].min()
        
        # Check if the amplitude exceeds the threshold and is the highest so far
        if amplitude > threshold and amplitude > max_amplitude:
            max_amplitude = amplitude
            enriched_metal = metal
    
    # Store the result for the current iModulon
    results[f'iModulon_{i}'] = enriched_metal if enriched_metal else 'No enrichment'
    data.loc[i, 'Enriched_Metal'] = enriched_metal if enriched_metal else 'No enrichment'
    
    
metal_columns = {
    'Cr': [f'Cr_{i}' for i in range(1, 7)],
    'Fe': [f'Fe_{i}' for i in range(1, 7)],
    'Cd': [f'Cd_{i}' for i in range(1, 7)],
    'Zn': [f'Zn_{i}' for i in range(1, 7)],
    'Cu': [f'Cu_{i}' for i in range(1, 7)],
    'Pb': [f'Pb_{i}' for i in range(1, 7)],
}

def calculate_center_of_mass(row):
    metal = row['Enriched_Metal']
    if metal in metal_columns:
        values = row[metal_columns[metal]].values
        indices = range(len(values))
        return sum(indices * values) / sum(values)
    else:
        return None

# Calculate the center of mass for each time series based on the enriched metal
data['center_of_mass'] = data.apply(calculate_center_of_mass, axis=1)

def calculate_highest_peak(row):
    metal = row['Enriched_Metal']
    if metal in metal_columns:
        values = row[metal_columns[metal]].values
        return np.argmax(values) + 1  # Adding 1 to make it 1-indexed
    else:
        return None

# Calculate the highest peak for each time series based on the enriched metal
data['highest_peak'] = data.apply(calculate_highest_peak, axis=1)


# Classify the shapes based on the combination of center of mass and highest peak
def classify_shapes(row):
    if row['center_of_mass'] is not None and row['highest_peak'] is not None:
        if row['center_of_mass'] <= 4 and row['highest_peak'] <= 3:
            return 'Fast Activation'
        elif row['center_of_mass'] <= 4 and row['highest_peak'] <= 4:
            return 'Intermediate Activation'
        elif row['center_of_mass'] <= 5 and row['highest_peak'] >= 5:
            return 'Steady Activation'
        else:
            return 'Mixed Activation'
    else:
        return 'No Activation'

data['shape_classification'] = data.apply(classify_shapes, axis=1)



# Select the last four columns of the dataset
selected_data = data.iloc[:, -4:]

# Create a new figure
fig, ax = plt.subplots(figsize=(12, 8))

# Hide axes
ax.axis('off')

# Create a table and add it to the figure
table_data = []
columns = selected_data.columns.insert(0, "iModulon")
table_data.append(columns)
for i, row in selected_data.iterrows():
    table_data.append([i] + row.tolist())

table = ax.table(cellText=table_data, cellLoc='center', loc='center')

# Adjust the table
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)

# Save the figure as a PNG file
plt.savefig('dataset_table.png', bbox_inches='tight')

# Optionally, display the figure
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the data
induction_dataset = ["Pb", "Zn", "Fe", "Cd", "Cr", "Pb", "Fe", "Zn", "Zn", "Zn", "Cu", "Cr", "Zn", "Fe", "Cu"]
post_induction_dataset = ["Cu", "Zn", "Zn", "Zn", "Zn", "Zn", "Fe", "Fe", "Cd", "Cu", "Zn", "Cu", "Pb", "Zn", "Cr", "Zn", "Zn", "Cr", "Fe", "Zn", "Cr", "Cr", "Cd", "Fe", "Fe", "Fe", "Zn", "Zn", "Zn", "Cr", "Fe", "Zn", "Cd", "Cd", "Pb"]

# Group the iModulons by heavy metal and get the iModulon numbers for each group
induction_groups = {metal: [i for i, x in enumerate(induction_dataset) if x == metal] for metal in set(induction_dataset)}
post_induction_groups = {metal: [i for i, x in enumerate(post_induction_dataset) if x == metal] for metal in set(post_induction_dataset)}

# Create bar plots
fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# Define a color map
colors = plt.cm.Paired(range(len(induction_groups)))

# Plot bar chart for induction dataset
bars1 = ax[0].bar(induction_groups.keys(), [len(v) for v in induction_groups.values()], color=colors)
ax[0].set_title('Induction Dataset')
ax[0].set_ylabel('Number of iModulons')
ax[0].set_xlabel('Heavy Metal')

# Plot bar chart for post-induction dataset
bars2 = ax[1].bar(post_induction_groups.keys(), [len(v) for v in post_induction_groups.values()], color=colors)
ax[1].set_title('Post-Induction Dataset')
ax[1].set_ylabel('Number of iModulons')
ax[1].set_xlabel('Heavy Metal')

# Add iModulon numbers as annotations within the bars
for i, bar in enumerate(bars1):
    metal = list(induction_groups.keys())[i]
    imodulons = 'iModulons: ' + ', '.join(map(str, induction_groups[metal]))
    ax[0].text(bar.get_x() + bar.get_width() / 2, bar.get_height() / 2, imodulons, ha='center', va='center', rotation=90, color='black', fontsize=10)

for i, bar in enumerate(bars2):
    metal = list(post_induction_groups.keys())[i]
    imodulons = 'iMod: ' + ', '.join(map(str, post_induction_groups[metal]))
    ax[1].text(bar.get_x() + bar.get_width() / 2, bar.get_height() / 2, imodulons, ha='center', va='center', rotation=90, color='black', fontsize=10)

# Display the plot
plt.tight_layout()
plt.savefig('iModulon_classification_barplot.png')  # Save the plot as a PNG file

plt.show()
```
